# Exercise 1.1 Self-Play

Suppose, instead of playing against a random opponent, the reinforcement learning algorithm  described above played against itself, with both sides learning. What do you think would happen in this case? Would it learn a different policy for selecting moves?

### Play 100 games of tic-tac-toe against itself

```python
ttt = tic_tac_toe()
results = ttt.play(100)
print(results)

# for player X, plot distribution of values not equal to 0, 0.5 or 1
plt.figure(figsize=(16, 5))
_ = plt.hist(ttt.value_fun_x[~np.isin(ttt.value_fun_x, [0, 0.5, 1])], bins=100)
plt.title('Value Function of Player X')
plt.show()

# for player O, plot distribution of values not equal to 0, 0.5 or 1
plt.figure(figsize=(16, 5))
_ = plt.hist(ttt.value_fun_o[~np.isin(ttt.value_fun_o, [0, 0.5, 1])], bins=100)
plt.title('Value Function of Player O')
plt.show()
```
